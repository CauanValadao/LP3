```python
class Data:
    def __init__(self, dia = 1, mes = 1, ano = 1, outraData = None, dias = 0):
        if not isinstance(outraData,Data):
            self._dia = dia
            self._mes = mes
            self._ano = ano
        else:
            self._dia = outraData._dia
            self._mes = outraData._mes
            self._ano = outraData._ano
            
        if not Data.valida(self._dia, self._mes, self._ano):
            self._dia = 1
            self._mes = 1
            self._ano = 1

        if dias != 0:
            self.somaData(dias)
        
    
    def valida(dia = 1, mes = 1, ano = 1, outraData = None):
        if isinstance(outraData, Data):
            dia = outraData.dia
            mes = outraData.mes
            ano = outraData.ano
        
        if ano < 0: return False
        if mes < 1 or mes > 12: return False
        if mes < 1 or dia > 31: return False
        if (mes == 4 or mes == 6 or mes == 9 or mes ==11) and dia > 30: return False
        if mes == 2:
            bissexto = False
            if(ano % 4 == 0 and ano % 100 != 0) or ano % 400 == 0: bissexto = True
            if dia > 29 or (dia == 29 and not bissexto): return False
        return True
    
    def __add__(self, dias):
        novaData = Data(outraData = self)
        novaData.somaData(dias = dias)
        return novaData
    
    def somaData(self, dias = 0, outraData = None):
        if isinstance(outraData, Data):
            self._dia = outraData.dia
            self._mes = outraData.mes
            self._ano = outraData.ano
            
        for i in range(dias):
                self._dia += 1
                if(not Data.valida(outraData = self)):  
                    self._dia = 1
                    self._mes += 1
                if(not Data.valida(outraData=self)):
                    self._mes = 1
                    self._ano += 1

    def __sub__(self, dias):
        novaData = Data(outraData=self)
        novaData.substraiData(dias = dias)
        return novaData

    def substraiData(self, dias=0, outraData=None):
        if isinstance(outraData, Data):
            self._dia = outraData.dia
            self._mes = outraData.mes
            self._ano = outraData.ano

        for _ in range(dias):
            self._dia -= 1
            if self._dia < 1:
                self._mes -= 1
                if self._mes < 1:
                    self._mes = 12
                    self._ano -= 1
                    if self._ano < 0:
                        self._ano = 0
                        self._mes = 1
                        self._dia = 1
                        break
                self._dia = 31
                while not Data.valida(self._dia, self._mes, self._ano):
                    self._dia -= 1

        


    def entraDatas(self):
        print("Digite uma data (dd/mm/aaaa):")
        self._dia = int(input("Digite o dia: "))
        self._mes = int(input("Digite o mes: "))
        self._ano = int(input("Digite o ano: "))
        if not Data.valida(outraData = self):
            print("Data invalida. Tente novamente")
            self.entraDatas()

    def mostraData(self):
        print(f'{self.dia:02d}/{self.mes:02d}/{self.ano:02d}')

    def __eq__(self, outraData):
        print("ola")
        if isinstance(outraData, Data):
            if(self.dia == outraData.dia and self.mes == outraData.mes and self.ano == outraData.ano):
                return True
            return False
        

    def igualA(data1, data2):
        return data1 == data2
    
    def __ne__(self, outraData):
        if isinstance(outraData, Data):
            if(self.dia != outraData.dia or self.mes != outraData.mes or self.ano != outraData.ano): 
                return True
            return False

    def diferenteDe(data1, data2):
        return data1 != data2
    
    def __lt__(self, outraData):
        if isinstance(outraData, Data):
            if self.ano > outraData.ano: return False
            elif self.mes > outraData.mes: return False
            elif self.dia >= outraData.dia: return False
            return True
        return False
    
    def menorQue(data1, data2):
        return data1 < data2
    
    def __gt__(self, outraData):
        if isinstance(outraData, Data):
            if self.ano < outraData.ano: return False
            elif self.mes < outraData.mes: return False
            elif self.dia <= outraData.dia: return False
            return True
        return False
    
    def maiorQue(data1, data2):
        return data1 > data2
    
    def __le__(self, outraData):
        return Data.menorQue(self, outraData) or Data.igualA(self, outraData)
    
    def menorOuIgual(data1, data2):
        data1 <= data2

    def __ge__(self, outraData):
        return Data.maiorQue(self, outraData) or Data. igualA(self, outraData)
    
    @property
    def dia(self):
        return self._dia
    
    @property
    def mes(self):
        return self._mes
    
    @property
    def ano(self):
        return self._ano
    
    @dia.setter
    def dia(self, dia):
        if Data.valida(dia, self.mes, self.ano):
            self.dia = dia
            return True
        return False
    
    @mes.setter
    def mes(self, mes):
        if Data.valida(self.dia, mes, self.ano):
            self.mes = mes
            return True
        return False

    @ano.setter
    def ano(self, ano):
        if Data.valida(self.dia, self.mes , ano):
            self.ano = ano
            return True
        return False
    


    def __str__(self):
        return f'{self.dia:02}/{self.mes:02}/{self.ano:04}'

```
